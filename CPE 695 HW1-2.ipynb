```python
##Problem2 a
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
def homework1(a,b,m):
    noise_scale=a
    number_of_samples=b
    x=25*(np.random.rand(number_of_samples,1)-0.8)
    y=5*x + 20*x**2 + 1*x**3 + noise_scale*np.random.randn(number_of_samples, 1)
    plt.plot(x,y,'ro')
    c=np.polyfit(x.flatten(), y.flatten(), m)
    p=np.poly1d(c)
    i = np.linspace(-20,5,1000)
    plt.plot(i,p(i))
    
homework1(100,50,5)
```

```python
##Problem2 b
def MSE(m):
    noise_scale=100
    number_of_samples=50
    x=25*(np.random.rand(number_of_samples,1)-0.8)
    y=5*x + 20*x**2 + 1*x**3 + noise_scale*np.random.randn(number_of_samples, 1)
    c=np.polyfit(x.flatten(), y.flatten(), m)
    p=np.poly1d(c)
    i = np.linspace(-20,5,1000)
    error=0
    for h in range(50):
        error+=(y[h]-p(x)[h])**2
    return np.mean(error)
```

```python
arr = []
for i in range(1,9):
    mse = MSE(i)
    arr.append(mse)
index = range(1,9)
plt.plot(index,arr)
```

The best order I found is 3.
